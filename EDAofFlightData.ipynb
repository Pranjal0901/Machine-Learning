```python
# Importing Basic libraries
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel('/content/sample_data/flight_price.xlsx')
df.head()
```

```python
# get basic info of dataset
df.info()
```

```python
# Feature Engineering
df['Date'] = df['Date_of_Journey'].str.split('/').str[0]
df['Month'] = df['Date_of_Journey'].str.split('/').str[1]
df['Year'] = df['Date_of_Journey'].str.split('/').str[2]
df.head()
```

```python
df['Date'] = df['Date'].astype(int)
df['Month'] = df['Month'].astype(int)
df['Year'] = df['Year'].astype(int)
```

```python
#Drop that column
df.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Arrival_Time'] = df['Arrival_Time'].apply(lambda x:x.split(' ')[0])
```

```python
df['Hour'] = df['Arrival_Time'].str.split(':').str[0]
df['Minutes'] = df['Arrival_Time'].str.split(':').str[1]
```

```python
df['Hour'] = df['Hour'].astype(int)
df['Minutes'] = df['Minutes'].astype(int)
```

```python
df.drop('Arrival_Time',axis=1,inplace=True)
```

```python
df['Dep_Hour'] = df['Dep_Time'].str.split(':').str[0]
df['Dep_Minutes'] = df['Dep_Time'].str.split(':').str[1]
```

```python
df['Dep_Hour'] = df['Dep_Hour'].astype(int)
df['Dep_Minutes'] = df['Dep_Minutes'].astype(int)
```

```python
df.head()
```

```python
df.drop('Dep_Time',axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Total_Stops'].mode()
```

```python
df['Total_Stops'].unique()
```

```python
df['Total_Stops']=df['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4,'Na':1})
```

```python
df.head()
```

```python
df[df['Total_Stops'].isnull()]
```

```python
df.loc[df['Total_Stops'].isna(), 'Total_Stops'] = 1
```

```python
df.head()
```

```python
df.drop('Route',axis=1,inplace=True)
```

```python
df.head()
```

```python
df['Duration_hour']=df['Duration'].str.split(' ').str[0].str.split('h').str[0]
df['Duration_min']=df['Duration'].str.split(' ').str[1].str.split('m').str[0]
df.loc[df['Duration_min'].isna(), 'Duration_min'] = 0
```

```python
df.head()
```

```python
df.drop('Duration',axis=1,inplace=True)
```

```python
df.head()
```

### Now we will convert these categorical data into numerical values with the help of One Hot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
Encoder = OneHotEncoder()
```

```python
Encoder.fit_transform(df[['Airline','Source','Destination']]).toarray()
```

```python
encode_df=pd.DataFrame(Encoder.fit_transform(df[['Airline','Source','Destination']]).toarray(),columns=Encoder.get_feature_names_out())
```

```python
df_cleaned = df.copy() # Create a copy to avoid modifying the original DataFrame if needed later
```

```python
final_df = pd.concat([df_cleaned.reset_index(drop=True), encode_df.reset_index(drop=True)], axis=1)
```
